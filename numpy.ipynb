## arange= array of range

```python
import numpy as np
a=np.array([1,2,3,4]) # creates an array
print(a)

print(np.arange(10)) # range function in arrray 
```

```python
L=range(1000)
square=[]

for i in range(1000):
    square.append(i**2)

print(square)

```

list comperhension

we can also write this code as

```python
L=range(1000)
square=[i**2 for i in L] # list comperhension
print(square)
```

only even numbers

```python
L=range(1000)
square=[i**2 for i in L if i % 2==0] # list comperhension
print(square)
```

 # creating arrays


```python
a=np.array([0,1,2,3])
print("array",a)
print("dimension",a.ndim)
print("shape",a.shape)
```

## 2d

```python
a=np.array([[0,2,3,4],[1,2,3,4]])
print("array",a)
print("dimension",a.ndim)
print("shape",a.shape)
```

## 3d

```python
a=np.array([[[0,2,3,4],[1,2,3,4]],[[0,1,2,3],[3,12,3,3]]])
print("array",a)
print("dimension",a.ndim)
print("shape",a.shape)
```
